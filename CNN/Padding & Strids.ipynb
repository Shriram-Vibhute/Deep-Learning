```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense, Conv2D, Flatten, Input
from keras import Sequential
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
# Returns the numpy nd array and not the dataframe
```

`valid padding` : *No padding is applied*

```python
# Building a sequential model
model = Sequential()

# Adding Layers
model.add(Input(shape = (28, 28, 1)))
model.add(Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu'))
model.add(Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu'))
model.add(Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu'))

# Adding Flatten Layer
model.add(Flatten())

# Adding Dense Layers
model.add(Dense(128, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))
```

```python
model.summary()
```

`same padding` : *Model will automatically configures the optimum number of paddings to set the image size before and after convolution*

```python
# Building a sequential model
model = Sequential()

# Adding Layers
model.add(Input(shape = (28, 28, 1)))
model.add(Conv2D(32, kernel_size=(3,3), padding='same', activation='relu'))
model.add(Conv2D(32, kernel_size=(3,3), padding='same', activation='relu'))
model.add(Conv2D(32, kernel_size=(3,3), padding='same', activation='relu'))

# Adding Flatten Layer
model.add(Flatten())

# Adding Dense Layers
model.add(Dense(128, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))
```

```python
model.summary()
```
